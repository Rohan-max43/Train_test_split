```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# read the dataset
iris_df = pd.read_csv("Iris.csv")
iris_df = iris_df.drop(columns=["Id"])
iris_df = iris_df.rename(columns={
    "SepalLengthCm":"sepal_length",
    "SepalWidthCm":"sepal_width",
    "PetalLengthCm":"petal_length",
    "PetalWidthCm":"petal_width",
    "Species":"species"
})
iris_df["species"].value_counts()
```

```python
# information about dataset
iris_df.info()
```

```python
# describe the dataset
iris_df.describe().round(5)
```

```python
# check is null values 
iris_df.isnull().sum()
```

```python
# histogram plot
iris_df["species"].value_counts().plot(kind="barh")
plt.show() 
```

```python
# Box plot with matplotlib
plt.figure(figsize=(8,6))
plt.boxplot(iris_df[["sepal_length","sepal_width","petal_length","petal_width"]], showmeans=True, sym="r*", vert=True, patch_artist=True, tick_labels=["sepal_length","sepal_width","petal_length","petal_width"])
plt.title("Box plot of Iris Dataset", fontsize=16)
plt.xlabel("Features", fontsize=14) 
plt.ylabel("cm", fontsize=14)
plt.grid()
plt.show()
```

```python
# box plot with seaborn 
plt.figure(figsize=(8,6))
sns.set_style("whitegrid")
sns.boxplot(data=iris_df[["sepal_length","sepal_width","petal_length","petal_width"]])
plt.title("Box plot of Iris Dataset", fontsize=16)
plt.xlabel("Features", fontsize=14)     
plt.ylabel("cm", fontsize=14)
plt.show()
```

```python
# boxen plot with seaborn 
plt.figure(figsize=(8,6))
sns.set_style("whitegrid")
sns.boxenplot(data=iris_df[["sepal_length","sepal_width","petal_length","petal_width"]])
plt.title("Box plot of Iris Dataset", fontsize=16)
plt.xlabel("Features", fontsize=14)     
plt.ylabel("cm", fontsize=14)
plt.show()
```

```python
# violin plot with seaborn
plt.figure(figsize=(8,6))
sns.set_style("whitegrid")
sns.violinplot(data=iris_df, x="species", y="sepal_length", hue="species")
plt.title("Violin plot of Sepal Length by Species", fontsize=16)
plt.xlabel("Species", fontsize=14)
plt.ylabel("Sepal Length (cm)", fontsize=14)
plt.show()
```

```python
# violin plot with matplotlib
plt.figure(figsize=(8,6))
plt.violinplot(dataset=iris_df[["sepal_length","sepal_width","petal_length","petal_width"]], vert=True, showmeans=True, showmedians=True)
plt.title("Violin plot of Sepal Length by Species", fontsize=16)
plt.xlabel("Species", fontsize=14)
plt.ylabel("Sepal Length (cm)", fontsize=14)
plt.show()
```

```python
# pair plot
plt.figure(figsize=(10,8))
sns.set_style("whitegrid")
sns.pairplot(iris_df[["sepal_length","sepal_width","petal_length","species"]], hue="species", diag_kind="kde")
plt.suptitle("Pair Plot of Iris Dataset", fontsize=16)
plt.show()
```

```python
g = sns.pairplot(iris_df[["sepal_length","sepal_width","petal_length","petal_width"]], diag_kind="kde")
g.map_lower(sns.kdeplot, levels=4, color=".2")
plt.show()
```

```python
# 2D scattter plot

sns.set_style("whitegrid")
sns.FacetGrid(iris_df, hue="species", height=4) \
    .map(plt.scatter, "sepal_length", "petal_width") \
    .add_legend()
plt.show()
```

```python
# 2D pair plot

sns.set_style("whitegrid")
sns.pairplot(iris_df, x_vars=["sepal_length","sepal_width"], y_vars=["petal_length","petal_width"], hue="species", height=5, aspect=1)
plt.suptitle("2D Pair Plot of Iris Dataset", fontsize=16)
plt.show()
```

```python
import plotly.express as px 

iris_df = px.data.iris()
fig = px.scatter_3d(iris_df, x="sepal_length", y="sepal_width", z="petal_width", color="species")
fig.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(iris_df[["sepal_length","sepal_width","petal_length","petal_width"]].iloc[:10],
            annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Iris Dataset", fontsize=16)           
plt.show()
```

```python
iris_df[["sepal_length","sepal_width","petal_length","petal_width"]].corr()
sns.heatmap(iris_df[["sepal_length","sepal_width","petal_length","petal_width"]].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset", fontsize=16)
plt.show()
```
